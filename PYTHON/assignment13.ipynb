    Two errors that are defined in the 'AirthmeticError' class are:
    1. 'ZeroDivisionError': This error is raised when trying to divide a number by zero.
    2. 'OverFlowError': This error is raised when a numerical calculation exceeds the maximum representable value.

```python
import logging
try:
    result = 1/0
except ZeroDivisionError:
    logging.exception("Cannot divide by zero")
```

```python
import logging

try:
    result = 9999990000 ** 999999999999999
except OverflowError:
    logging.exception("Number too large to calculate")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

    The 'LookupError' class in python is a base class for exception that occur when a specified key or index cannot be found in a sequence or mapping object. This class is typically used to handle errors that occur when looking up values in a sequnce or mapping object.
    Two common exceptions that inherit from the LookupError class in Python are KeyError and IndexError.
    A KeyError exception is raised when a key is not found in a dictionary or other mapping object. For example:

```python
my_dict = {'foo':42,'bar':23}
try:
    value = my_dict['baz']
except KeyError:
    print("Key not found ")
```

    This code creates a dictionary my_dict with two key-value pairs, 'foo': 42 and 'bar': 23. It then attempts to access a key 'baz' that does not exist in the dictionary.
    Since the key is not found, it raises a KeyError exception. The code catches this exception using a try-except block, with the except clause handling the KeyError. 

```python
my_list = [1,2,3]
try:
    value: my_list[3]
except IndexError:
    print("Index out of range")
```

    In this example, the code attempts to look up the value at index 3 in the list my_list. However, since the  list  
    only has 3 elements (with indices 0, 1, and 2), an IndexError is raised. The try block catches the exception, and 
    the code in the except block is executed, which simply prints an error message.

    In both cases, the try block attempts to look up a value, and the except block handles the LookupError that is  
    raised if the value is not found. By using the LookupError class, the code can handle both KeyError and    
    IndexError exceptions with a single except block.

Q5. Explain ImportError. What is ModuleNotFoundError?

        ImportError is a built-in Python exception that is raised when an imported module, package, or object cannot be found or loaded. This can occur for a variety of reasons, such as when the module is misspelled, not installed, or is not in the correct directory.

    ModuleNotFoundError is a subclass of ImportError that was introduced in Python 3.6. It is raised when a module cannot be found during import, and is more specific than the generic ImportError. The purpose of this subclass is to make it easier to catch and handle import errors that are caused specifically by a missing module.
    Here's an example that demonstrates how ImportError and ModuleNotFoundError can be raised:

```python
try:
    import non_existent_module
except ImportError as e:
    print(f"ImportError: {str(e)}")

try:
    import non_existent_module_2
except ModuleNotFoundError as e:
    print(f"ModuleNotFoundError: {str(e)}")
```

Q6. List down some best practices for exception handling in python.

        Here are some best practices for exception handling in Python:

    1) Use specific exception types whenever possible instead of catching all exceptions with a broad try- 
    except block.
    2) Only catch exceptions that you know how to handle, otherwise let them propagate up the call stack.
    3) Avoid using bare except: statements, as they catch all exceptions, including system exit signals and 
    keyboard interrupts, which could make it difficult to diagnose and debug issues.
    4) Use finally blocks to ensure that resources are properly released, even if an exception is thrown.
    5)Use context managers, such as the with statement, to automatically handle the opening and closing of 
    resources.
    6)Log exceptions with as much detail as possible, including the stack trace, to aid in debugging and    
    troubleshooting.
    7)Handle exceptions close to the source of the problem, as opposed to higher up in the call stack.
    When re-raising exceptions, use the raise statement without any arguments to preserve the original 
    traceback.
