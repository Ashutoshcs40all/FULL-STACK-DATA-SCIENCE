Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a 
scatter plot for age and fare columns in the titanic dataset.


```python
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
import cv2
```

```python
titanic=sns.load_dataset("titanic")
```

```python
fig = px.scatter(titanic, x="age", y="fare",color="age")
fig.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in 
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" 
column with the color parameter.

```python
tips = px.data.tips() 
fig = px.histogram(tips,x="sex", y="total_bill",pattern_shape="smoker",color="day")
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for 
the color parameter

        Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the 
dimensions parameter.

```python
iris=px.data.iris()
fig=px.scatter_matrix(iris,dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],color="species")
fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

        Distplot (short for distribution plot) is a type of plot used to visualize the distribution of a single variable. It is a combination of a histogram and a kernel density plot, showing the frequency distribution of the variable as well as an estimate of its probability density function.

```python
import plotly.express as px
tips = px.data.tips()
fig = px.histogram(tips, x="total_bill", nbins=30, opacity=0.5, color_discrete_sequence=["blue"])
fig.update_layout(title="Distribution of Total Bill", xaxis_title="Total Bill", yaxis_title="Count")
fig.show()
```
