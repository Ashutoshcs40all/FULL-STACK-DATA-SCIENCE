Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.

        A scatter plot is used to display data points as individual points on a graph, and is commonly used to show the relationship between two variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt 

plt.scatter(x,y)
plt.xlabel("This is x Axis")
plt.ylabel("This is y Axis")
plt.title("This is random Graph")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
        Use the following data:

        The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to divide a single figure into multiple subplots and plot different datasets or visualizations side-by-side, making it easier to compare them.
        The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot. The index starts at 1 and increases row-wise, meaning that the first subplot is located in the top-left corner and subsequent subplots are added to the right of the previous one until the row is full, at which point the next row is filled.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])               #line 1 DATA
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])                       #line2 DATA
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])                           #line 3 DATA
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])                               #line 4 DATA
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')                                            # Plot line 1 on the first subplot

axs[0, 1].plot(x2, y2)                                  # Plot line 2 on the second subplot
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)                                              # Plot line 3 on the third subplot
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)                                          # Plot line 4 on the fourth subplot
axs[1, 1].set_title('Line 4')

fig.suptitle('Four Line Plots')
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

        A bar plot, also known as a bar chart, is a type of graph that presents categorical data with rectangular bars, with the height or length of each bar representing the numerical value of the data. The bars can be oriented horizontally or vertically.

        Bar plots are used to visually compare different categories of data and show the relative size or frequency of each category. They are particularly useful when dealing with discrete or categorical data, where there are a limited number of distinct categories. Bar plots can show both individual values as well as aggregated values such as means, medians, or sums of values. They are also useful for identifying trends or patterns in the data.

        Bar plots are commonly used in various fields such as business, finance, social sciences, and public health to display and compare data. For example, a bar plot can be used to compare the sales performance of different products in a company, or to compare the prevalence of different diseases across different age groups in a population. Bar plots are easy to read, understand, and interpret, which makes them an effective way of presenting data to a wider audience.






        



```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Company Name")
plt.ylabel("Company Profit")
plt.title("MNC Company Data")
plt.show
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

        A box plot, also known as a box-and-whisker plot, is a type of graphical representation of statistical data that displays the distribution of a dataset. It consists of a box that spans the interquartile range (IQR) of the data, with a vertical line inside the box representing the median value of the data. The "whiskers" extend from the edges of the box to show the range of the data, excluding outliers, which are plotted as individual points outside the whiskers.
        Box plots are used to visually summarize and compare the distribution of a dataset or to identify outliers. They provide a quick and simple way to examine the spread, central tendency, and variability of the data, and to compare the distribution of multiple datasets side by side. Box plots are especially useful for datasets with a large number of observations or when comparing multiple datasets with different sample sizes or units of measurement.
        Box plots are widely used in various fields such as statistics, data science, finance, and social sciences. For example, in finance, box plots can be used to display the distribution of stock prices or returns, and to identify potential outliers or extreme values. In medical research, box plots can be used to display the distribution of clinical trial data, and to compare the effectiveness of different treatments or interventions. Overall, box plots provide a simple yet effective way to summarize and compare datasets, making them a valuable tool for data analysis and visualization.



```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
fig, ax = plt.subplots()
ax.boxplot(data)
plt.xlabel("Dataset")
plt.ylabel("Value")
plt.title("Box Plot of Two Datasets")
plt.show()
```
